```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('train.csv')
df_numeric = df.select_dtypes(include=['number'])
df_numeric = df_numeric[df_numeric.columns[1:]]
sns.heatmap(df_numeric.corr(), cmap='coolwarm', annot=True, fmt='.2f')
plt.show()
```

```
correlation_matrix = df_numeric.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix,
            mask=mask,
            annot=False,
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={'shrink': 0.8})

plt.title('HeatMap - Dataset', fontsize=14, pad=15)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

```
n_features = len(df_numeric.columns)
fig_size = max(12, n_features * 0.6)  

plt.figure(figsize=(fig_size, fig_size))
sns.heatmap(correlation_matrix, 
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={'shrink': 0.8},
            annot_kws={'size': 8})  

plt.title('HeatMap - Dataset', fontsize=14, pad=15)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.show()
```
